```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
rd = pd.read_pickle(r"C:\Users\kalya\Desktop\jupyter\TMY3_datafile.pkl")
rd['Site identifier code'] = rd['Site identifier code'].astype('category')
rd['Station name'] = rd['Station name'].astype('category')
rd['Station state'] = rd['Station state'].astype('category')
rd['Site latitude'] = rd['Site latitude'].astype('category')
rd['Site longitude'] = rd['Site longitude'].astype('category')
rd['Site elevation'] = rd['Site elevation'].astype('category')
```

```python
rd.rename(columns={'Date (MM/DD/YYYY)':'Dt','Time (HH:MM)':'Time'})
```
